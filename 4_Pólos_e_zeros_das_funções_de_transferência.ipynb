```python
#importa as bibliotecas
import control as ctl
import matplotlib.pyplot as plt
import numpy as np
```

```python
#cria a função de transferência
P_s = ctl.tf([1,2],[1,1])
print('FT=',P_s)
```

```python
#calcula os pólos e zeros
ps_ma = ctl.pole(P_s)
zs_ma = ctl.zero(P_s)
#mostra os pólos e zeros
print('polos=',ps_ma)
print('zeros=',zs_ma)
```

```python
#faz o diagrama pz
plt.figure()
ctl.pzmap(P_s,1,0,'diagrama pz de P_s')
```

```python
#cria a função de transferência
P_s = ctl.tf([1,1],[1,8,15])
print('FT=',P_s)
#calcula os pólos e zeros
ps_ma = ctl.pole(P_s)
zs_ma = ctl.zero(P_s)
#mostra os pólos e zeros
print('polos=',ps_ma)
print('zeros=',zs_ma)
#faz o diagrama pz
plt.figure()
ctl.pzmap(P_s,1,0,'diagrama pz de P_s')
```

```python
#cria a função de transferência
R_s = ctl.tf([1],[1,0])
print('R_s=',R_s)
P_s = ctl.tf([1,2],[1,1])
print('P_s=',P_s)
C_s = R_s*P_s
#calcula os pólos e zeros
ps_ma = ctl.pole(C_s)
zs_ma = ctl.zero(C_s)
#mostra os pólos e zeros
print('polos de C_s=',ps_ma)
print('zeros de C_s=',zs_ma)
#faz o diagrama pz
plt.figure()
ctl.pzmap(C_s,1,0,'diagrama pz de C_s')
```

```python
#resposta ao degrau
Tsim=10
T,yout = ctl.step_response(P_s,Tsim)
#calcula um degrau unitário
Tdegrau = np.linspace(-1,Tsim,1000)
degrau = np.ones_like(Tdegrau)
degrau[Tdegrau<0]=0
#plota os resultados
plt.figure()
plt.plot(T,yout,'k-')
plt.plot(Tdegrau,degrau,'r-')
plt.xlabel('tempo(s)')
plt.legend(['resposta ao degrau','degrau unitário'])
plt.grid()
```
