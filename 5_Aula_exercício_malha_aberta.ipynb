```python
import control as ctl
import matplotlib.pyplot as plt
import numpy as np
```

```python
R = 0.5
L = 1e-6
C = 1e-6
T_sim = 2.5e-5

P_q = ctl.tf([1/L], [1, R/L, 1/(L*C)])
P_i = ctl.tf([1/L, 0], [1, R/L, 1/(L*C)])
P_vc = ctl.tf([1/(L*C)], [1, R/L, 1/(L*C)])

T_q, y_q = ctl.step_response(P_q, T_sim)
T_i, y_i = ctl.step_response(P_i, T_sim)
T_vc, y_vc = ctl.step_response(P_vc, T_sim)

plt.figure()
plt.plot(T_q, y_q, 'b-')
plt.xlabel('tempo (s)')
plt.legend(['carga no capacitor'])
plt.grid()

plt.figure()
plt.plot(T_i, y_i, 'r-')
plt.xlabel('tempo (s)')
plt.legend(['corrente no circuito'])
plt.grid()

plt.figure()
plt.plot(T_vc, y_vc, 'k-')
plt.xlabel('tempo (s)')
plt.legend(['tensão no capacitor'])
plt.grid()
```
